```python
# Importing Libraries

import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns

plt.style.use("ggplot")

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (10,6)


# read the data

df = pd.read_csv(r"C:\Users\Poornima\Desktop\projects\Python_Movies\archive\movies.csv")
```

```python
df.head()
```

```python
# Looking for missing data
    
df.isnull().sum() 
```

```python
df.dropna(how = "any").shape
```

```python
df.shape
```

```python
#Dropping rows if any of the value is null. 70% of the data is retained
df = df.dropna()
```

```python
# Data types of columns
df.dtypes
```

```python
# Changing data type of columns
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')
df['runtime'] = df['runtime'].astype('int64')
```

```python
df.head()
```

```python
# Creating a new column year_correct by getting the correct year from the 'released' column
#df = df.drop('year_correct')
df['year_correct'] = df['released'].str.split(expand=True)[2]
df.head()
```

```python
# Sorting based on gross
df.sort_values(by = ['gross'], inplace = False, ascending = False)
```

```python
# Checking for duplicates and dropping them

df.drop_duplicates()
```

```python
# Correlation of Budget with gross revenue
# Correlation of Company with gross revenue
```

```python
# Scatter plot of Budget Vs Gross

plt.scatter(df['budget'],df['gross'])
plt.title("Budget Vs Gross Earnings")
plt.xlabel("Budget")
plt.ylabel("Gross Earnings")
```

```python
sns.regplot(x = "budget", y ="gross", data = df, scatter_kws = {"color":"brown"}, line_kws = {"color":"black"})
```

```python
# Looking at the correlation 

df.corr(method = "pearson")
```

```python
# High correlation between Budget and Gross

corr_matrix = df.corr(method = "pearson")
sns.heatmap(corr_matrix, annot = True)
plt.title("Correlation matric for numeric features")
plt.xlabel("Movie features")
plt.ylabel("Movie features")
```

```python
# Correlation between company and gross
# Making all the columns with data type as 'object' into coded categorical variables
```

```python
df_numerical = df

for cols in df_numerical.columns:
    if df_numerical[cols].dtype == 'object':
        df_numerical[cols] = df_numerical[cols].astype('category')
        df_numerical[cols] = df_numerical[cols].cat.codes
```

```python
df_numerical
```

```python
corr_matrix = df_numerical.corr(method = "pearson")
sns.heatmap(corr_matrix, annot = True)
plt.title("Correlation matric for numeric features")
plt.xlabel("Movie features")
plt.ylabel("Movie features")
```

```python
corr_matrix = df_numerical.corr()
corr_matrix
```

```python
corr_pairs = corr_matrix.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.50]
high_corr
```
